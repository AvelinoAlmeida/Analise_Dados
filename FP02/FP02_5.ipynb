```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/vinhoVerde.csv')
```

```python
# Encode the 'tipo' column into binary values (tinto=0, branco=1)
df['tipo_encoded'] = df['tipo'].apply(lambda x: 1 if x == 'branco' else 0)

```

```python
# Features remain the same, but now the target variable is the encoded 'tipo'
X = df.drop(columns=['quality', 'tipo', 'tipo_encoded'])  # Exclude 'quality' and 'tipo'
y = df['tipo_encoded']
```

```python
# Split the dataset: 75% for training, 25% for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

```

```python
# Train a decision tree classifier
decision_tree_clf = DecisionTreeClassifier(random_state=42)
decision_tree_clf.fit(X_train, y_train)
```

```python
# Predict on the test set
y_pred_clf = decision_tree_clf.predict(X_test)
```

```python
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred_clf)
```

```python
# Visualize the decision tree
plt.figure(figsize=(20, 10))
tree.plot_tree(decision_tree_clf, feature_names=X.columns, class_names=['Tinto', 'Branco'], filled=True)
plt.title("Decision Tree for Predicting Wine Type (Tinto/Branco)")
plt.show()
```

```python
# Return accuracy result
accuracy
```

O modelo de árvore de decisão treinado para prever o tipo de vinho (tinto ou branco) obteve uma accuracy de aproximadamente 98% nos dados de teste, o que significa que o modelo acertou na classificação em cerca de 98% das instâncias.

Interpretação dos resultados:
Árvore de Decisão: A árvore de decisão mostra quais características do vinho são mais importantes para distinguir entre vinhos tintos e brancos. Por exemplo, variáveis como "ácidos voláteis", "densidade" e "sulfitos" tendem a ser determinantes na separação dos dois tipos de vinho.

Alto desempenho: A alta taxa de acerto (accuracy) sugere que as características químicas e físicas dos vinhos são bastante distintas entre vinhos tintos e brancos, permitindo ao modelo fazer previsões com elevada confiança.
